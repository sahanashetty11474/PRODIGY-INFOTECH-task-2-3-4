```python
# TASK1
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
titanic_df = pd.read_csv(r"C:\Users\sahan\Downloads\train.csv")
```

```python
# Display the first few rows of the dataset
print(titanic_df.head())
```

```python
# Data Cleaning

# Check for missing values
missing_values = titanic_df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
# Drop the Cabin column due to high number of missing values
titanic_df = titanic_df.drop(columns=['Cabin'])

```

```python
# Convert categorical variables to category type
titanic_df['Sex'] = titanic_df['Sex'].astype('category')
titanic_df['Embarked'] = titanic_df['Embarked'].astype('category')

```

```python
# Fill missing values in Age with the median age
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())
```

```python
# Fill missing values in Embarked with the mode (most frequent value)
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])
```

```python
# Verify that there are no more missing values
print("Missing values after cleaning:\n", titanic_df.isnull().sum())
```

```python
# Exploratory Data Analysis (EDA)

# Survival rate
survival_rate = titanic_df['Survived'].mean()
print(f"Overall survival rate: {survival_rate:.2f}")
```

```python
# Survival rate by gender
survival_rate_gender = titanic_df.groupby('Sex', observed=True)['Survived'].mean()
print("Survival rate by gender:\n", survival_rate_gender)
```

```python
# Survival rate by passenger class
survival_rate_pclass = titanic_df.groupby('Pclass', observed=True)['Survived'].mean()
print("Survival rate by passenger class:\n", survival_rate_pclass)
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

# Survival rate by age
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df[titanic_df['Survived'] == 1]['Age'], bins=30, kde=True, color='blue', label='Survived')
sns.histplot(titanic_df[titanic_df['Survived'] == 0]['Age'], bins=30, kde=True, color='red', label='Not Survived')
plt.title('Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
 #Fare distribution
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# Survival rate by fare
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df[titanic_df['Survived'] == 1]['Fare'], bins=30, kde=True, color='blue', label='Survived')
sns.histplot(titanic_df[titanic_df['Survived'] == 0]['Fare'], bins=30, kde=True, color='red', label='Not Survived')
plt.title('Survival Rate by Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Embarkation points
embarkation_counts = titanic_df['Embarked'].value_counts()
print("Passenger distribution by embarkation points:\n", embarkation_counts)
```

```python
# Survival rate by embarkation points
survival_rate_embarked = titanic_df.groupby('Embarked', observed=True)['Survived'].mean()
print("Survival rate by embarkation points:\n", survival_rate_embarked)
```

```python

# Visualization: Survival rate by gender and class
plt.figure(figsize=(10, 6))
sns.catplot(x='Pclass', hue='Sex', col='Survived', data=titanic_df, kind='count', height=4, aspect=0.7)
plt.show()
```
